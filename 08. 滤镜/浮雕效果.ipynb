```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('fa.jpg')
```

```python
#自定义卷积核
kernel_emboss_1 = np.array([[0,-1,-1],[1,0,-1],[1,1,0]])
kernel_emboss_2 = np.array([[1,1,0],[-1,0,1],[0,-1,-1]])
kernel_emboss_3 = np.array([[1,0,0],[0,0,0],[0,0,-1]])

#浮凸
output_1 = cv2.filter2D(img,-1,kernel_emboss_1)
output_2 = cv2.filter2D(img,-1,kernel_emboss_2)
output_3 = cv2.filter2D(img,-1,kernel_emboss_3)
```

```python
plt.subplot(131)
plt.imshow(output_1)
plt.subplot(132)
plt.imshow(output_2)
plt.subplot(133)
plt.imshow(output_3)
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
gray = cv2.imread('fa.jpg')
```

```python
imgInfo = gray.shape
```

```python
height, weight, ch = imgInfo
```

```python
dst = np.zeros((height,weight,1),np.uint8)
```

```python
for i in range(0,height):
    for j in range(0,weight-1):
        gray0=gray[i,j]
        gray1=gray[i,j+1]
        newp=gray0-gray1+150
        if newp[0]+newp[1]+newp[2]>255:
            newp=255
        else:
            newp=0
        dst[i,j]=newp
```

```python
plt.imshow(dst)
```
